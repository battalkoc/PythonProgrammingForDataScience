##while döngüsü

```python
i=1
while i<10:
  print(i)
  i+=1
```

```python
x=10
while x<13:
  print(x)
  x+=1
```

```python
for i in range(10,13):
  print(i)
```

```python
i=1
while i<6:
  print(i)
  if i==3:
    break
  i+=1
```

```python
i=0
while i<6:
  i+=1
  if i==3:
    continue
  print(i)
```

```python
i=1
while i<6:
  print(i)
  i+=1
else:
  print("bu bukadar")
```
