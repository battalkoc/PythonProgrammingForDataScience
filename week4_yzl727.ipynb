##Arbitrary Arguments, *args

```python
def func(*kids):
  print("The youngest child is", kids[2])
```

```python
func("ali","veli","ahmet","ayse", "murat", "mehmet", "mustafa")
```

##Keyword Arguments

key=value

```python
def girls(girl1, girl2, girl3):
  print("The youngest girl is " + girl1)
```

```python
girls(girl1="Ayse", girl2="Fatma", girl3="Hayriye")
```

##Arbitrary Keyword Arguments, **kwargs

```python
children(name="Murat", surname="simsek")
```

```python
def passing(midterm, final):
  meanMidterm=midterm*0.4
  meanFinal=final*0.6
  meanGeneral=meanMidterm+meanFinal
  if meanGeneral>45 and meanGeneral<60:
    print("Genel Ortalama: ", meanGeneral)
    print("şartlı geçtin")
  elif meanGeneral>=60:
    print("Genel Ortalama: ", meanGeneral)
    print("Geçtin")
  else:
    print("Genel Ortalama: ", meanGeneral)
    print("Kaldın")
```

```python
midterm=int(input("ara sınav notunu giriniz:"))
final=int(input("final notunu giriniz:"))
```

```python
passing(midterm, final)
```

```python
def calculationFactorial():
  number=int(input("faktoriyel alınacak sayısı giriniz:"))
  factorial=1

  if number>=0:
    for i in range(1, number+1):
      factorial=factorial*i
    print("factorial:", factorial)
  else:
    print("negatif sayı girmeyiniz")
```

```python
calculationFactorial()
```
