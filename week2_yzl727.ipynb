and operator

```python
a=200
b=33
c=500

if a>b and c>a:
  print("a büyüktür b ve c de büyüktür a")
```

or operator

```python
a=200
b=33
c=500

if a>b or a>c:
  print("a değeri b deperinden büyüktür veya adeğeri c deperinden büyüktür")
```

```python
x=41

if x>10:
  print("10 dan büyük")
  if x>20:
    print("ayrıca 20den de büyük")
  else:
    print("10 dan büyük ama 20 den kücük ")
```
