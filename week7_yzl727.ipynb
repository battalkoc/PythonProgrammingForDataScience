#**Overriding**

Method overriding is an ability of any object-oriented programming language that allows a subclass or child class to provide a specific implementation of a method that is already provided by one of its super-classes or parent classes. When a method in a subclass has the same name, same parameters or signature and same return type(or sub-type) as a method in its super-class, then the method in the subclass is said to override the method in the super-class.



Geçersiz kılma, bir sınıfın, temel sınıflarından biri tarafından sağlanan bir yöntemin uygulamasını değiştirme özelliğidir.

Geçersiz kılma, kalıtımın tüm gücünü kullanmasını sağladığı için OOP'nin çok önemli bir parçasıdır. Bir sınıfı geçersiz kılma yöntemini kullanarak, başka bir sınıfı "kopyalayabilir", yinelenen kodlardan kaçınabilir ve aynı zamanda onun bir bölümünü geliştirebilir veya özelleştirebilir. Yöntem geçersiz kılma bu nedenle kalıtım mekanizmasının bir parçasıdır.

Python yönteminde geçersiz kılma, alt sınıfta, üst sınıftaki bir yöntemle aynı ada sahip bir yöntem tanımlanarak gerçekleşir. Nesnede bir yöntem tanımladığınızda, bu yöntemin bu yöntem çağrısını karşılayabilmesini sağlarsınız, böylece atalarının uygulamaları devreye girmez.

```python
#Parent Class

class Person:

  #constructor

  def __init__(self,name, id):
    self.name=name
    self.id=id


  def display(self):
    print(self.name, self.id)
```

```python
class Sirket:
  def __init__(self, isim):
    print("Şirket sınıfı oluşturuldu")
    self.isim=isim

  def bolumler(self):
    print("şirket bölümleri")

  def idare(self):
    print("müdür")
```

```python
sirket=Sirket("Murat")
```

```python
class Parent:

  def __init__(self):
    print("parent")
    
  def display(self):
    print("Inside Parent")

class Child(Parent):
  def show(self):
    print("Inside Child")

class GrandChild(Child):
  def show(self):
    print("Inside GrandChild")
```

```python
g=GrandChild()
```

```python
from abc import ABC,abstractmethod

class Polygon(ABC):

  @abstractmethod
  def numberofsides(self):
    pass
    
```

```python
class Triangle(Polygon):
  def numberofsides(self):
    print("I have 3 sides")
```

```python
class Hexagon(Polygon):
  def numberofsides(self):
    print("I have 4 sides")
```

```python
class Quadrilateral(Polygon):
  def numberofsides(self):
    print("I have 6 sides")
```

```python
class Pentagon(Polygon):
  def numberofsides(self):
    print("I have 5 sides")
```

```python
t=Triangle()
```

```python
t.numberofsides()
```

```python
p=Pentagon()
```

```python
p.numberofsides()
```
